# Deleting rows and columns by using $drop()$ function

#The $drop()$ function in Pandas be used to delete rows and columns from a DataFrame with the axis set to 0(row) or set to 1(column).

#inplace parameter can be used to alter DataFrames without reassignment.

```python
data.drop('Unnamed: 0',axis=1)  # here unnamed column is deleted,but it is temperarly
```

```python
data # see again column unnamed is present in data, because we deleted temperarly
```

```python
data.drop('Unnamed: 0',axis=1,inplace=True) #here unnamed data columns is deleted permanently because 'inplaced=True' is used
```

```python
data #see here unnamed column is deleted permanently
```

#deleting row if it unwanted suppose, exmaple taking row 1

```python
data.drop(1,axis=0) # here row 1 is deleted but  it is temperarly
```

```python
data #see again row 1  is present in data, because we deleted temperarly
```

```python
data.drop(1,axis=0,inplace=True) ##here row 1 data is deleted permanently because 'inplaced=True' is used
```

```python
data   #see here row 1 is deleted permanently
```

# Filling missing values using fillna(), replace()

```python
# filling missing value using fillna(), 
df.fillna(0) # here total data will be fill with 0 in place of NaN
```

```python
#Filling null values with the previous ones
```

```python
df.fillna(method='ffill')  # or you can use method='pad'
```

```python
Df = pd.DataFrame([[1,np.nan,2,None],
                  [2,3,5,np.nan],
                  [np.nan,4,9,None],
                  [10,20,30,40]],columns=['A','B','C','D'])
Df
```

```python
Df.dropna(axis=1) #by default you get column which does not have null values
```

```python
Df.dropna(axis='columns',how='all') #we will get all values including null values
```
