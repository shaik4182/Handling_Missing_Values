# Checking for missing values using following Methods

#$isna()$ and $notnull()$:

#In order to check missing values in Pandas DataFrame, we use a function $isnull()$ and $notnull()$.Both function help in checking whether a value is NaN or not.These function can also be used in Pandas Series in order to find null values in a Series

#method=1

#using $isna()$ and $isnull()$ means,Method creates a Boolean Series for extracting rows with null values (True=null values, False=not null values)

```python
data.isna() # or use data.isnull()
```

#Method =2

#using $notnull()$ and $notna()$ means, means,Method creates a Boolean Series for extracting rows with non null values (True= not null values, False=null values)

```python
data.notna() # or use data.notnull()
```
